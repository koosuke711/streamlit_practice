```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tickers = {
    'apple': 'AAPL',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'netflix': 'NFLX',
    'amazon': 'AMZN'
}

days = 20
```

```python
tickers['apple']
```

```python
tkr = yf.Ticker(tickers['apple'])
hist = tkr.history(period='20d')
hist.index = hist.index.strftime('%d %B %Y')
```

```python
def get_data(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])
    return df
```

```python
get_data(days, tickers)
```

```python
import altair as alt
```

```python
df = get_data(days, tickers)
df
```

```python
df_ex = df.loc[['apple', 'google']]
```

```python
df_ex = df_ex.T
```

```python
df_ex = df_ex.reset_index()
```

```python
df_ex = pd.melt(df_ex, id_vars=['Date']).rename(columns={'value': 'Stock Prices'})
df_ex
```

```python
ymin, ymax = 175, 192
chart = (
    alt.Chart(df_ex)
    .mark_line(opacity=0.8, clip=True)
    .encode(
        x='Date:T',
        y=alt.Y('Stock Prices:Q', stack=None, scale=alt.Scale(domain=[ymin, ymax])),
        color='Name:N'
    )
)
```
